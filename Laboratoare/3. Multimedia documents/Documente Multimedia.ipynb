#### Citirea unui fisier XML
In exemplul de mai jos, va citi fisierul XML *adnotare1.xml* si se va extrage din ele pozitia obiectelor in imaginea cu acelasi nume.
Dupa extragere, se vor afisa patratele care incadreaza obiectele din imagine si denumirea obiectelor.

```python
import xml.etree.ElementTree as ET
import cv2

file_name = "adnotare1.xml"

image_name = file_name.split('.')[0]
img = cv2.imread(image_name + '.jpeg')
print(image_name)
tree = ET.parse(file_name)
root = tree.getroot()

for obj in root.iter('object'):
    name = obj.find('name').text
    print(name)
    for box in obj.iter('bndbox'):
        xmin = int(box.find('xmin').text)
        ymin = int(box.find('ymin').text)
        xmax = int(box.find('xmax').text)
        ymax = int(box.find('ymax').text)
        print(xmin, ymin, xmax, ymax)

        img = cv2.rectangle(img, (xmin, ymin), (xmax, ymax), color=(255, 0, 0), thickness=2)
        img = cv2.putText(img, name, (xmin, ymin), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)

cv2.imshow("Imagine adnotata", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## DOCX

Pentru a lucra cu fisiere de tip DOCX, vom folosi biblioteca python-docx, a carei documentatie o puteti gasi la: https://python-docx.readthedocs.io/en/latest/.

Este posibil ca unele functionalitati din Microsoft Word sa nu fie disponibile in biblioteca, insa in unele cazuri acestea se pot defini pe baza functiilor deja existente. Acest lucru se intampla si in functia **add_hyperlink**, care este construita pentru a introduce un hyperlink marcat corespunzator pe un text dintr-un paragraf.

```python
# Referinta: https://github.com/python-openxml/python-docx/issues/74
def add_hyperlink(paragraph, text, url):
    """
    A function that places a hyperlink within a paragraph object.

    :param paragraph: The paragraph we are adding the hyperlink to.
    :param url: A string containing the required url
    :param text: The text displayed for the url
    :return: A Run object containing the hyperlink
    """

    # This gets access to the document.xml.rels file and gets a new relation id value
    part = paragraph.part
    r_id = part.relate_to(url, docx.opc.constants.RELATIONSHIP_TYPE.HYPERLINK, is_external=True)

    # Create the w:hyperlink tag and add needed values
    hyperlink = docx.oxml.shared.OxmlElement('w:hyperlink')
    hyperlink.set(docx.oxml.shared.qn('r:id'), r_id )
    hyperlink.set(docx.oxml.shared.qn('w:history'), '1')

    # Create a w:r element
    new_run = docx.oxml.shared.OxmlElement('w:r')

    # Create a new w:rPr element
    rPr = docx.oxml.shared.OxmlElement('w:rPr')

    # Create a w:rStyle element, note this currently does not add the hyperlink style as its not in
    # the default template, I have left it here in case someone uses one that has the style in it
    rStyle = docx.oxml.shared.OxmlElement('w:rStyle')
    rStyle.set(docx.oxml.shared.qn('w:val'), 'Hyperlink')

    # Join all the xml elements together add add the required text to the w:r element
    rPr.append(rStyle)
    new_run.append(rPr)
    new_run.text = text
    hyperlink.append(new_run)

    # Create a new Run object and add the hyperlink into it
    r = paragraph.add_run()
    r._r.append(hyperlink)

    # A workaround for the lack of a hyperlink style (doesn't go purple after using the link)
    # Delete this if using a template that has the hyperlink style in it
    r.font.color.theme_color = docx.enum.dml.MSO_THEME_COLOR_INDEX.HYPERLINK
    r.font.underline = True

    return r
```

#### Genrare fisier DOCX

```python
import docx
from docx.shared import Inches

document = docx.Document()

# Titlu document
document.add_heading('Aplicatii Multimedia Laborator 3', 0)

p = document.add_paragraph('Acest document va contine informatii ')
p.add_run('diverse (scris bold)').bold = True
p.add_run(' despre ')
p.add_run('programare (scris italic)').italic = True
p.add_run(' si ')
p.add_run('inignerie (scris italic).').italic = True

document.add_heading('Capitolul 1', level=1)
document.add_paragraph('Cateva notiuni de electronica', style='Intense Quote')

p = document.add_paragraph('Elemente pasive de circuit')
document.add_paragraph( 'Rezistente', style='List Bullet' )
document.add_paragraph( 'Condensatoare', style='List Bullet' )
document.add_paragraph( 'Bobine', style='List Bullet' )

p = document.add_paragraph('Tipuri de motoare')
document.add_paragraph( 'De curent continuu (DC) cu perii', style='List Number' )
document.add_paragraph( 'De curent continuu fara perii', style='List Number' )
document.add_paragraph( 'Steppere', style='List Number' )
document.add_paragraph( 'Servo motoare', style='List Number' )

document.add_heading('Cele mai folosite limbaje de programare', level=2)

tabel = document.add_table(rows=1, cols=3)
hdr_cells = tabel.rows[0].cells
hdr_cells[0].text = 'Denumire'
hdr_cells[1].text = 'Loc 2020'
hdr_cells[2].text = 'Loc 2019'

data = ( ('Javascript', 1, 1),
         ('Java',       2, 2),
         ('C',          3, 3),
         ('Python',     4, 4) )

for den, loc2020, loc2019 in data:
    celule_rand = tabel.add_row().cells
    celule_rand[0].text = den
    celule_rand[1].text = str(loc2020)
    celule_rand[2].text = str(loc2019)

p = document.add_paragraph('Raportul complet poate fi gasit ')
hyperlink = add_hyperlink(p, text='aici', url='https://info.hackerrank.com/rs/487-WAY-049/images/HackerRank-2020-Developer-Skills-Report.pdf')

document.add_page_break()
    
document.add_heading('Cum folosim imagini', level=3)
document.add_picture('comic.png', width=Inches(2.5))

document.save('demo.docx')
```

## Fisiere PDF

#### Generara unui fisier PDF care contine un tabel si cateva imagini

```python
from fpdf import FPDF

left_margin = 20
right_margin = 20
top_margin = 20

# Definim o clasa care mosteneste FPDF si fpdf.HTMLMixin
# pentru a putea folosi etichete HTML pentru formatarea textului
# Referinta: https://github.com/reingart/pyfpdf/blob/master/docs/reference/write_html.md
class FPDF_Class(FPDF, fpdf.HTMLMixin):
    pass

pdf = FPDF_Class()

pdf.set_margins(left=left_margin, top=top_margin, right=right_margin)
print("Latime efectiva PDF (in inch): ", pdf.epw)

pdf.add_page()
# Setare font 
pdf.set_font('Arial', 'B',size=20)
pdf.cell(0, 5, 'Aplicatii Multimedia Laborator 3', ln=1)
# Trasarea unei linii din marginea din stanga pana la marginea din dreapta
pdf.line(x1=left_margin, y1=30, x2=pdf.epw+left_margin, y2=30)
pdf.ln()

pdf.set_font("Times", 'B', size=12)
# Scrierea unui paragraf folosind etichete HTML
pdf.write_html("<p>Acest document va contine informatii <b>diverse (scris bold)</b> despre <i>programare (scris italic)</i> si <i>inignerie (scris italic)</i>.</p>")
pdf.ln(20)


pdf.set_font('Arial', 'B',size=20)
pdf.cell(0, 5, 'Capitolul 1', ln=1)
pdf.set_font("Times", 'B', size=12)
# Lista neordonata folosind HTML
pdf.write_html("<p>Elemente pasive de circuit:\
                <ul><li>Rezistente</li><li>Condensatoare</li><li>Bobine</li></ul></p>")
pdf.ln()
# Lista ordonata folosind HTML
pdf.write_html("<p>Tipuri de motoare:\
                <ol><li>De curent continuu (DC) cu perii</li>\
                <li>De curent continuu fara perii</li>\
                <li>Steppere</li>\
                <li>Servo motoare</li></ol></p>")
pdf.ln(20)


pdf.set_font('Arial', 'B',size=20)
pdf.cell(0, 5, 'Cele mai folosite limbaje de programare', ln=1)
pdf.ln(5)

data = ( ('Denumire', 'Loc 2020', 'Loc 2019'),
         ('Javascript', '1', '1'),
         ('Java',       '2', '2'),
         ('C',          '3', '3'),
         ('Python',     '4', '4') )

pdf.set_font("Times", size=12)
# Setare intaltime rand pentru a incadra textul
line_height = pdf.font_size * 2.5
# Setare latime pentru distribuirea coloanelor in mod egal
col_width = pdf.epw / 4 
# Scriere celule tabel
for row in data:
    for datum in row:
        pdf.multi_cell(col_width, line_height, datum, border=1, ln=3, max_line_height=pdf.font_size)
    pdf.ln(line_height)
pdf.ln(5)
# Adaugare Link
pdf.cell(w=40, h=10, txt="Raportul complet poate fi gasit aici.", link="https://info.hackerrank.com/rs/487-WAY-049/images/HackerRank-2020-Developer-Skills-Report.pdf")

pdf.add_page()
pdf.set_font('Arial', 'B',size=20)
pdf.cell(0, 5, 'Undeva la un laborator in Poli', ln=1)
pdf.image( "confused.jpeg", w=125 )
pdf.ln(5)

pdf.output('demo.pdf')
```

## Fisiere CSV

#### Crearea unui fisier CSV in Python

```python
import csv  
import numpy as np

header = ['Valoare1', 'Valoare2', 'Suma']
x = np.random.rand(10)
y = np.random.rand(10)
suma = x + y

with open('date.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)

    # Scriere header
    writer.writerow(header)

    # Scriere date
    for i in range(0, len(x)):
        writer.writerow([x[i], y[i], suma[i]])
```
