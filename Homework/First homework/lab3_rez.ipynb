```python
##### CIUCIU ANCA-MARIA - 333 AA #####

##### GENERARE XML #####

import xml.etree.ElementTree as ET
import cv2

xmin = 593
ymin = 24
xmax = 689
ymax = 369

img = cv2.imread('adnotare2.jpeg')
img = cv2.rectangle(img, (xmin, ymin), (xmax, ymax), color=(255, 0, 0), thickness=2)
img = cv2.putText(img, 'Door', (xmin, ymin), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)

cv2.imshow("Imagine", img)
cv2.waitKey(0)
cv2.destroyAllWindows()

nume_fisier = 'adnotare2.xml'

root = ET.Element("annotation")

e1 = ET.Element("filename")
e1.text = "img"
root.append(e1)

e2 = ET.Element("size")
root.append(e2)

s1 = ET.SubElement(e2, "width")
s1.text = "1248"
s2 = ET.SubElement(e2, "height")
s2.text = "384"
s3 = ET.SubElement(e2, "depth")
s3.text = "3"

e3 = ET.Element("object")
root.append (e3)

s1 = ET.SubElement(e3, "name")
s1.text = "Door"

tree = ET.ElementTree(root)

with open(nume_fisier, "wb") as files :
    tree.write(files)


##### GENERARE DOCX #####

import docx
from docx.shared import Inches
import numpy as np
from matplotlib import pyplot as plt

document = docx.Document()

# Titlu document
document.add_heading('Ciuciu Anca-Maria Tema Aplicatii Multimedia', 0)
document.add_heading("Tabel",1)

x = np.sort(np.random.rand(10))
y = x * x + 10

plt.plot(x, y, ls='solid', linewidth=2, color='black')
plt.title("Grafic functie")
plt.savefig("graficEx2.png")

tabel = document.add_table(rows=1, cols=3)
hdr_cells = tabel.rows[0].cells
hdr_cells[0].text = 'x'
hdr_cells[1].text = 'y'

for i in range(10):
    celule_rand = tabel.add_row().cells
    celule_rand[0].text = str(x[i])
    celule_rand[1].text = str(y[i])

document.add_heading("Grafic",1)
plt.plot(x, y, ls='solid', linewidth=2, color='black')
plt.savefig("generated_plot_docx.png")
document.add_picture("generated_plot_docx.png")

document.save('Output.docx')


##### GENERARE PDF #####

import fpdf
from fpdf import FPDF
import csv
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

header = ['Valoare1', 'Valoare2', 'Suma']
x = np.sort(np.random.rand(10))
y = np.sort(np.random.rand(10))
suma = x + y


with open('date.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)

    # Scriere header
    writer.writerow(header)

    # Scriere date
    for i in range(0, len(x)):
        writer.writerow([x[i], y[i], suma[i]])

xr = []
yr = []
sumr = []

with open('date.csv', newline='\n') as f:
    reader = csv.reader(f)
    for row in reader:
        xr.append(row[0])
        yr.append(row[1])
        sumr.append(row[2])

print(f"Xr este: {xr}")
print(f"Yr este: {yr}")
print(f"Sumr este: {sumr}")


data = []

for i in range(len(xr)):
    data.append((xr[i], yr[i], sumr[i]))

left_margin = 20
right_margin = 20
top_margin = 20

class FPDF_Class(FPDF, fpdf.HTMLMixin):
    pass

pdf = FPDF_Class()
pdf.set_margins(left=left_margin, top=top_margin, right=right_margin)

pdf.add_page()
# Setare font
pdf.set_font('Arial', 'B',size=20)
pdf.cell(0, 5, 'Ciuciu Anca-Maria Tema Aplicatii Multimedia', ln=1)
pdf.ln(20)
pdf.set_font('Arial', 'B',size=20)
pdf.cell(0, 5, 'Tabel', ln=1)
pdf.ln(10)
pdf.set_font("Times", 'B', size=12)

pdf.set_font("Times", size=12)
line_height = pdf.font_size * 2.5
col_width = pdf.epw / 4

for row in data:
    for i in row:
        pdf.multi_cell(col_width, line_height, i, border=1, ln=3, max_line_height=pdf.font_size)
    pdf.ln(line_height)
pdf.ln(5)

pdf.add_page()

pdf.set_font('Arial', 'B',size=20)
pdf.cell(0, 5, 'Grafic', ln=1)
plt.plot(x, y, ls='solid', linewidth=2, color='black')
plt.savefig("generated_plot_pdf.png")
pdf.image( "generated_plot_pdf.png", w=125)
pdf.ln(5)


pdf.output('Output.pdf')
```
